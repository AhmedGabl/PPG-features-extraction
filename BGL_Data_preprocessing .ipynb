```python
#get rid of inconsistent values 

import pandas as pd
from sklearn.preprocessing import MinMaxScaler

# Read the data from the Excel file
data = pd.read_excel('BGL_Data/pr_PG2.xlsx')
data.columns = data.columns.str.strip()

# Select relevant columns for analysis
selected_columns = ['Age', 'Heart Rate', 'Diastolic Blood Pressure', 'Systolic Blood Pressure', 'SPO2', 'Blood Glucose Level(BGL)']
preprocessed_data = data[selected_columns].copy()

# Remove rows with inconsistent output values for the same input features
preprocessed_data = preprocessed_data.drop_duplicates(subset=['Age', 'Heart Rate', 'Diastolic Blood Pressure', 'Systolic Blood Pressure', 'SPO2'], keep=False)

# Handling missing values
preprocessed_data = preprocessed_data.dropna()

# Save the preprocessed data back to an Excel file
preprocessed_data.to_excel('BGL_Data/pr_pr_PG2.xlsx', index=False)
```

```python

# Data normalization using Min-Max scaling
scaler = MinMaxScaler()
preprocessed_data[selected_columns[:-1]] = scaler.fit_transform(preprocessed_data[selected_columns[:-1]])

# Save the preprocessed data back to an Excel file
preprocessed_data.to_excel('BGL_Data/pr_pr_PG2.xlsx', index=False)
```
